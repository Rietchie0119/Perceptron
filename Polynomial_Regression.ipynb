```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```python
np.random.seed(0)
points = 500
x = np.linspace(-3, 3, points)
y = np.sin(x) + np.random.uniform(-0.5, 0.5, points)
plt.scatter(x, y)
```

```python
model = Sequential()
model.add(Dense(50, activation='sigmoid', input_dim=1))
model.add(Dense(30, activation='sigmoid'))
model.add(Dense(1))
```

```python
adam = Adam(learning_rate=0.01)
model.compile(loss='mse', optimizer=adam) # mean square error 은 실제 y값과 fitting line의 예상 값의 차를 비교하는 기법. 차를 구한뒤 제곱한것을 다 더함. 마지막으로 n 으로 나눈다.
# gradient 는 가장 기울기가 경사진 곳을 가리키는 벡터. Cost function의 derivative 라고 보면 됨. 현재의 정의역x(weight 또는 bias)에서 gradient의 역방향 벡터와 임의의 learning rate를 곱한 쪽으로 가면서 
# gradient descent즉 최적의 fitting을 향새 간다고 볼수 있음.
model.fit(x, y, epochs=50)
```

```python
predictions = model.predict(x)
plt.scatter(x, y)
plt.plot(x, predictions, 'ro') # 빨간색 동그라미로 predictions를 보여줌
plt.show()
```
